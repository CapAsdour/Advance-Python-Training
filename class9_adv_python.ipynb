```python
# Function in panda
import pandas as pd
import numpy as np
def add(e1,e2):
    return e1+e2
df=pd.DataFrame(np.random.randn(5,3),columns=['col1','col2','col3'])
df.pipe(add,2)
print(df.apply(np.mean))
```

```python
import pandas as pd
import numpy as np
def add(e1,e2):
    return e1+e2
df=pd.DataFrame(np.random.randn(5,3),columns=['col1','col2','col3'])
df.pipe(add,2)
print(df.apply(np.mean,axis=1))
```

```python
n=20
df=pd.DataFrame({'A':pd.date_range(start='2019-06-06',periods=n,freq='D'),'X':np.linspace(0,stop=n-1,num=n),'Y':np.random.rand(n),'C':np.random.choice(['Low','Medium','High'],n).tolist(),'D':np.random.normal(200,10,size=(n)).tolist()})
df
```

```python
for col in df:
    print(col)
```

```python
for key,value in df.iteritems():
    print(key,value)
```

```python
for row_index,row in df.iterrows():
    print(row_index,row)
```

```python
for row in df.itertuples():
    print(row)
```

```python
st=df.sort_index()
print(st)
```

```python
st=df.sort_index(ascending=False)
print(st)
```

```python
st=df.sort_index(axis=1)
print(st)
```

```python
st=df.sort_index(by='Y')
print(st)
```

```python
st=df.sort_index(by='Y',kind='mergesort')
print(st)
```
