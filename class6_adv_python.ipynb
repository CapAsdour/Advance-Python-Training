```python
import matplotlib.pyplot as plt
import numpy as np
fig=plt.figure(figsize=(20,10))
x=[22.2,17.6,8.8,8,7.7]
y=['tiger','deer','impala','elephant','bear']
color=["red","lightgreen","lightblue","green","yellow"]
explodes=(0.1,0,0,0,0)
plt.pie(x,explode=explodes,labels=y,colors=color,autopct='%1.1f%%',shadow=True,startangle=140)
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig,ax=plt.subplots()
size=0.3
vals=np.array([[30.,32.],[37.,40.],[29.,10.]])
cmap=plt.get_cmap("tab20c")
outer_colors=cmap(np.arange(3)*4)
inner_colors=cmap(np.array([1,2,5,6,9,10]))
ax.pie(vals.sum(axis=1),radius=3,colors=outer_colors,wedgeprops=dict(width=3*size,edgecolor='w'))
ax.pie(vals.flatten(),radius=3-size,colors=inner_colors,wedgeprops=dict(width=3*size,edgecolor='w'))
ax.set(aspect="equal",title="plot")
plt.show()
```

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
#setting a style
style.use('ggplot')
#create a new fig
fig=plt.figure()
ax1=fig.add_subplot(111,projection='3d')
x=np.random.randint(0,10,size=20)
y=np.random.randint(0,10,size=20)
z=np.random.randint(0,10,size=20)
#plot
ax1.plot(x,y,z)
ax1.set_xlabel('x axis')
ax1.set_ylabel('y axis')
ax1.set_zlabel('z axis')
plt.show()
```

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
#setting a style
style.use('ggplot')
#create a new fig
fig=plt.figure()
ax1=fig.add_subplot(111,projection='3d')
u,v=np.mgrid[0:2*np.pi:200j,0:np.pi:100j]
x=np.cos(u)*np.sin(v)
y=np.sin(u)*np.sin(v)
z=np.cos(v)
#plot
ax1.plot_wireframe(x,y,z,rstride=5,cstride=5,linewidth=1)
ax1.set_xlabel('x axis')
ax1.set_ylabel('y axis')
ax1.set_zlabel('z axis')
plt.show()
```
